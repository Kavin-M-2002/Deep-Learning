```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(10,6)
```

```python
digits_df=keras.datasets.mnist
```

```python
(x_train_full,y_train_full),(x_test,y_test)=digits_df.load_data()
```

```python
plt.imshow(x_train_full[0])
```

```python
y_train_full
```

```python
x_test
```

```python
y_test
```

```python
#Data Reshape
x_train_fulls=x_train_full.reshape((60000,28,28,1))
x_tests=x_test.reshape((10000,28,28,1))
```

```python
x_train_fulls
```

```python
x_tests
```

```python
#Data Normalization
x_train_n=x_train_full/255.
x_test_n=x_test/255.
```

```python
#Splitting the data
x_valid,x_train=x_train_n[:6000],x_train_n[6000:]
y_valid,y_train=y_train_full[:6000],y_train_full[6000:]
```

```python
#creating model architecture
np.random.seed(42)
tf.random.set_seed(42)
```

```python
#pooling
model_pool=keras.models.Sequential()
model_pool.add(keras.layers.Conv2D(filters=32,kernel_size=(3,3),strides=1,padding='valid',activation='relu',input_shape=(28,28,1)))
model_pool.add(keras.layers.MaxPooling2D((2,2)))
model_pool.add(keras.layers.Flatten())
model_pool.add(keras.layers.Dense(300,activation="relu"))
model_pool.add(keras.layers.Dense(100,activation="relu"))
model_pool.add(keras.layers.Dense(10,activation="softmax"))

#non-pooling
model_non_pool=keras.models.Sequential()
model_non_pool.add(keras.layers.Conv2D(filters=32,kernel_size=(3,3),strides=1,padding='valid',activation='relu',input_shape=(28,28,1)))
model_non_pool.add(keras.layers.Flatten())
model_non_pool.add(keras.layers.Dense(200,activation="relu"))
model_non_pool.add(keras.layers.Dense(100,activation="relu"))
model_non_pool.add(keras.layers.Dense(10,activation="softmax"))
```

```python
print(model_pool.summary())
print(model_non_pool.summary())
```

```python
model_pool.compile(loss="sparse_categorical_crossentropy",
             optimizer="sgd",
              metrics=['accuracy'])

model_non_pool.compile(loss="sparse_categorical_crossentropy",
             optimizer="sgd",
              metrics=['accuracy'])
```

```python
model_history_pool=model_pool.fit(x_train,y_train,epochs=60,batch_size=64,
                                 validation_data=(x_valid,y_valid))
```

```python
model_history_non_pool=model_non_pool.fit(x_train,y_train,epochs=60,batch_size=64,
                                 validation_data=(x_valid,y_valid))
```

```python
model_history_pool.history
```

```python
model_history_non_pool.history
```

```python
pd.DataFrame(model_history_pool.history).plot(figsize=(10,6))
plt.grid(True)
import seaborn as sns
sns.set_style('darkgrid')
plt.gca().set_ylim(0,1)
plt.title('Pooling')
plt.show()
```
